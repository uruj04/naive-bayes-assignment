```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB
from sklearn.preprocessing import Binarizer
from sklearn.metrics import accuracy_score

iris_data = load_iris()
features = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
labels = iris_data.target

X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.3, random_state=42
)

gaussian_classifier = GaussianNB()
gaussian_classifier.fit(X_train, y_train)

g_train_acc = accuracy_score(y_train, gaussian_classifier.predict(X_train))
g_test_acc = accuracy_score(y_test, gaussian_classifier.predict(X_test))

print(f"Gaussian Model")
print(f"Training: {g_train_acc:.3f} ; Test: {g_test_acc:.3f}\n")

mean_cutoff = features.mean().mean()
data_simplifier = Binarizer(threshold=mean_cutoff)

X_train_binary = data_simplifier.fit_transform(X_train)
X_test_binary = data_simplifier.transform(X_test)

bernoulli_classifier = BernoulliNB()
bernoulli_classifier.fit(X_train_binary, y_train)

b_train_acc = accuracy_score(y_train, bernoulli_classifier.predict(X_train_binary))
b_test_acc = accuracy_score(y_test, bernoulli_classifier.predict(X_test_binary))

print(f"Bernoulli Model")
print(f"Training: {b_train_acc:.3f} ; Test: {b_test_acc:.3f}")
```
